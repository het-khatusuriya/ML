## <font color="maroon"><h4 align="center">Handling Missing Data - fillna, interpolate, dropna</font>

```python
import pandas as pd
df = pd.read_csv("weather_data.csv")
#type(df.day[0])
#here we read the dataframe
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
#type(df.day[0])
df
```

```python
df.set_index('day',inplace=True)
#here we set the days as index
df
```

## <font color="blue">fillna</font>

<font color="purple">**Fill all NaN with one specific value**</font>

```python
#here we can fil the NULL value as 0
new_df = df.fillna(0)
new_df
```

<font color="purple">**Fill na using column names and dict**</font>

```python
new_df = df.fillna({
    #if temp and wind are 0 the there are no events
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

<font color="purple">**Use method to determine how to fill na values**</font>

```python
new_df = df.fillna(method="ffill")
#from that methos we fill data of upper row
new_df
```

```python
new_df = df.fillna(method="bfill")
#from that methods we fill data of lower row

new_df
```

<font color="purple">**Use of axis**</font>

```python
new_df = df.fillna(method="bfill", axis="columns") # axis is either "index" or "columns"
#from that methods we fill data of right column

new_df
```

<font color="purple">**limit parameter**</font>

```python
new_df = df.fillna(method="ffill",limit=1)
#from that methods we fill data of left column

new_df
```

### <font color="blue">interpolate</font>

```python
new_df = df.interpolate()
#from that method we can find value between two point
new_df
```

```python
# Create a new DataFrame, 'new_df', by performing time-based interpolation on the existing DataFrame 'df'

new_df = df.interpolate(method="time")

# Display the resulting interpolated DataFrame
new_df
```

**Notice that in above temperature on 2017-01-04 is 29 instead of 30 (in plain linear interpolate)**

**There are many other methods for interpolation such as quadratic, piecewise_polynomial, cubic etc.
Just google "dataframe interpolate" to see complete documentation**

### <font color="blue">dropna</font>

```python
new_df = df.dropna()
#from that we can drop the NULL value row
new_df
```

```python
new_df = df.dropna(how="all")
new_df
#from that we can drop the NULL value row

print(df)
```

```python
new_df = df.dropna(thresh=2)
new_df
```

### <font color="blue">Inserting Missing Dates</font>

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
#from that we can fill the missing date from "01-01-2017","01-11-2017" and set the days as index
df.reindex(idx)
```
